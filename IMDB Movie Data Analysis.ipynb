## Groupby


### What are the mean ratings of first 5 directors ?

```python
data.groupby('Director')[['Rating']].mean().head() 
```

### Who are the top 5 highest rated directors and what are their ratings ?

```python

data.groupby('Director')[['Rating']].mean().sort_values(['Rating'],ascending=False).head() 
```

### Apply() Function

Let's add a rating category column to the data.

```python
def rating_group(rating):
    if rating>=7.5 :
        return 'Good'
    elif rating>=6.0 :
        return 'Average'
    else :
        return 'Bad'

```

```python
data['Rating_category']=data['Rating'].apply(rating_group)
```
